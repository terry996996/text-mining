```python
with open('ntusd-positive.txt', encoding = 'utf-8', mode = 'r') as f:
  positive_words = []
  for l in f:
    positive_words.append(l.strip())
```

```python
with open('ntusd-negative.txt', encoding = 'utf-8', mode = 'r') as f:
  negative_words = []
  for l in f:
    negative_words.append(l.strip())
```

```python
score = 0
jieba_result = jieba.cut(article, cut_all=False, HMM=True)
for word in jieba_result:
  if word in positive_words:
    score += 1
  elif word in negative_words:
    score -= 1
  else:
    pass
  print(f'詞彙:{word}, 總分:{score}')
```

```python
import pandas as pd
```

```python
stopword_list = [k.strip() for k in open('stopword.txt', encoding = 'utf-8').readlines() if k.strip() != ""]
```

```python
print(stopword_list)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# 使用cut_for_search(搜尋引擎)斷詞模式並產生字詞頻率的dictionary (並剔除stopwords的計算)
Sentence = jieba.cut_for_search(d)

# create a python dictionary
hash = {}
for item in Sentence:

    if item in stopword_list:
        continue

    if item in hash:
        hash[item] += 1
    else:
        hash[item] = 1

# 文字雲樣式設定
wc = WordCloud(font_path="NotoSansCJKtc-Regular.otf", #設置字體
               background_color="white", #背景顏色
               max_words = 2000 ,        #文字雲顯示最大詞數
               stopwords=stopword_list)      #停用字詞

# 使用dictionary的內容產生文字雲
wc.generate_from_frequencies(hash)

# 視覺化呈現
plt.imshow(wc)
plt.axis("off")
plt.figure(figsize=(20,10), dpi =200)
plt.show()
```

```python
Sentence = jieba.cut_for_search(d)
text = ' '.join(Sentence)

# 文字雲樣式設定
wc = WordCloud(font_path="NotoSansCJKtc-Regular.otf", #設置字體
               background_color="white", #背景顏色
               max_words = 2000 ,        #文字雲顯示最大詞數
               stopwords=stopword_list)      #停用字詞

# 直接使用結巴斷詞後的文字來產生文字雲
wc.generate_from_text(text)

# 視覺化呈現
plt.imshow(wc)
plt.axis("off")
plt.figure(figsize=(20,10), dpi =200)
plt.show()
```

```python
artDf = pd.DataFrame.from_dict(hash, orient='index', columns = ['詞頻'])
abc = artDf.sort_values(by= ['詞頻'], ascending= False)
abc.to_csv('output.csv', encoding = 'utf-8-sig')
```
